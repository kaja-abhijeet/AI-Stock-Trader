````python
from pypdf import PdfReader

reader = PdfReader("Data Input.pdf")
text = ""

for page in reader.pages:
    text += page.extract_text() + "\n"

print(text[:1000])
````

````python
prompt = f"""
Your task is to convert the PDF text into structured key:value rows.

!!! STRICT REQUIREMENTS !!!

1. For EVERY extracted row, you MUST output:
   - "Key"
   - "Value"
   - "Comments"

2. "Comments" MUST:
   - Include any contextual information related to the key/value.
   - Preserve original wording.
   - NOT be empty unless absolutely no context exists.
   - Include multi-line context when needed.

3. You MUST capture 100% of the PDF content.
   Nothing can be skipped or summarized.

4. Output MUST be ONLY a valid Python list of dictionaries in this exact format:
   [
      {{"Key": "...", "Value": "...", "Comments": "..."}},
      {{"Key": "...", "Value": "...", "Comments": "..."}},
      ...
   ]

5. Absolutely DO NOT:
   - Add markdown
   - Add backticks
   - Add explanations

Here is the input text:
{text}
"""
````

````python
from openai import OpenAI
client = OpenAI(api_key=OPENAI_API_KEY)

response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[{"role": "user", "content": prompt}],
)

structured_output = response.choices[0].message.content
print(structured_output)
````

````python
import ast
import pandas as pd


cleaned_output = (
    structured_output.replace("```python", "")
    .replace("```", "")
    .strip()
)

data = ast.literal_eval(cleaned_output)

df = pd.DataFrame(data)
df.to_excel("Output.xlsx", index=False)

print("Output.xlsx created successfully!")
````
